```python
from langchain_text_splitters import HTMLHeaderTextSplitter


html_string="""
<!DOCTYPE html>
<html>
  <head>
    <title>Sample Content Page</title>
  </head>
  <body>
    <h1>Welcome to My Website</h1>
    <p>This is a demo page to showcase headings and paragraph content.</p>

    <h2>About This Page</h2>
    <p>This page contains different levels of headings: h1, h2, and h3. It's useful for demonstrating content hierarchy in HTML structure.</p>

    <h3>Why Use Headings?</h3>
    <p>Headings help organize content, improve accessibility, and boost SEO by providing structure to web pages.</p>
  </body>
</html>
"""

header_tosplit=[("h1","header1"),("h2","header2"),("h3","header3")]
html_splitter=HTMLHeaderTextSplitter(header_tosplit)
htmls=html_splitter.split_text(html_string)
print(htmls)
```

```python
url="https://en.wikipedia.org/wiki/Large_language_model"

html2=html_splitter.split_text_from_url(url)
print(html2)
```
